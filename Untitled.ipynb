```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import tensorflow as tf
from keras.models import Sequential, Model
from keras.layers import Dense, Conv2D, Dropout, Flatten, MaxPooling2D, Input
from PIL import Image
from tqdm.notebook import tqdm
from keras.utils import load_img
from keras import backend as K
from tensorflow.keras.utils import plot_model
```

```python
images=[]
ages=[]
genders=[]
```

```python
dir=os.listdir('data_set')
for filename in (dir):
    image_path=os.path.join('data_set',filename)
    temp=filename.split('_')
        
    try:
        age=int(temp[0])
    except:
        continue
    try:
        gender=int(temp[1])
        if(gender>1):
            continue
    except:
        continue
        
    
    images.append(image_path)
    genders.append(gender)
    ages.append(age)
    
    

    
```

```python
df = pd.DataFrame()
df['image'], df['age'], df['gender'] = images, ages, genders
df.head()
```

```python
gender_dict = {0:'Male', 1:'Female'}
```

```python
def extract_features(images):
    features = []
    for image in tqdm(images):
        img = load_img(image, grayscale=True)
        img = img.resize((128, 128))
        img = np.array(img)
        features.append(img)
        
    features = np.array(features)
    # ignore this step if using RGB
    features = features.reshape(len(features), 128, 128, 1)
    return features
```

```python
X = extract_features(df['image'])
```

```python
X.shape
```

```python
X = X/255.0
```

```python
y_gender = np.array(df['gender'])
y_age = np.array(df['age'])
```

```python
def age_accuracy_with_tolerance(y_true, y_pred):
    # Extract true and predicted ages
    true_age = y_true
    pred_age = y_pred

    # Define tolerance level
    tolerance = 3

    # Check if the predicted age is within the tolerance range
    within_tolerance = K.abs(true_age - pred_age) <= tolerance

    # Convert boolean values to float (1 for True, 0 for False) and calculate mean
    accuracy = K.mean(K.cast(within_tolerance, dtype='float32'))

    return accuracy
```

```python
input_shape = (128, 128, 1)
```

```python
inputs = Input((input_shape))
# convolutional layers
conv_1 = Conv2D(32, kernel_size=(3, 3), activation='relu') (inputs)
maxp_1 = MaxPooling2D(pool_size=(2, 2)) (conv_1)
conv_2 = Conv2D(64, kernel_size=(3, 3), activation='relu') (maxp_1)
maxp_2 = MaxPooling2D(pool_size=(2, 2)) (conv_2)
conv_3 = Conv2D(128, kernel_size=(3, 3), activation='relu') (maxp_2)
maxp_3 = MaxPooling2D(pool_size=(2, 2)) (conv_3)
conv_4 = Conv2D(256, kernel_size=(3, 3), activation='relu') (maxp_3)
maxp_4 = MaxPooling2D(pool_size=(2, 2)) (conv_4)

flatten = Flatten() (maxp_4)

# fully connected layers
dense_1 = Dense(256, activation='relu') (flatten)
dense_2 = Dense(256, activation='relu') (flatten)

dropout_1 = Dropout(0.3) (dense_1)
dropout_2 = Dropout(0.3) (dense_2)

output_1 = Dense(1, activation='sigmoid', name='gender_out') (dropout_1)
output_2 = Dense(1, activation='relu', name='age_out') (dropout_2)

model = Model(inputs=[inputs], outputs=[output_1, output_2])

model.compile(loss=['binary_crossentropy', 'mae'], optimizer='adam', metrics=['accuracy',age_accuracy_with_tolerance])
```

```python

plot_model(model)
```

```python
history = model.fit(x=X, y=[y_gender, y_age], batch_size=32, epochs=30, validation_split=0.2)
```

```python
model.save('completemodel2.h5')
```

```python
from keras.models import load_model

loaded_model = load_model('completemodel2.h5',custom_objects={'age_accuracy_with_tolerance': age_accuracy_with_tolerance})
```

```python
acc = history.history['gender_out_accuracy']
val_acc = history.history['val_gender_out_accuracy']
epochs = range(len(acc))

plt.plot(epochs, acc, 'b', label='Training Accuracy')
plt.plot(epochs, val_acc, 'r', label='Validation Accuracy')
plt.title('Accuracy Graph')
plt.legend()
plt.figure()

loss = history.history['gender_out_loss']
val_loss = history.history['val_gender_out_loss']

plt.plot(epochs, loss, 'b', label='Training Loss')
plt.plot(epochs, val_loss, 'r', label='Validation Loss')
plt.title('Loss Graph')
plt.legend()
plt.show()
```

```python
loss = history.history['age_out_loss']
val_loss = history.history['val_age_out_loss']
epochs = range(len(loss))

plt.plot(epochs, loss, 'b', label='Training Loss')
plt.plot(epochs, val_loss, 'r', label='Validation Loss')
plt.title('Loss Graph')
plt.legend()
plt.show()
```
